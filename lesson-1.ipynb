```python
def note(pitch_adjust, time):
    x = (list(range(0,round(100/pitch_adjust))))*int(440*pitch_adjust*time/3)
    x = x[:-44]+[0]*44
    return(x)
```

```python
def bassnote(pitch_adjust, time):
    return(note(pitch_adjust*1/8,time))
```

```python
def bassline():
    bassline = bassnote(1,1)
    bassline.extend(bassnote(1,1))
    bassline.extend(bassnote(2,1))
    bassline.extend(bassnote(7/4,1))
    bassline.extend(bassnote(2,1))
    bassline.extend(bassnote(4/3,.5))
    bassline.extend(bassnote(3/2,1))
    bassline.extend(bassnote(7/4,.5))
    bassline.extend(bassnote(2,1)) 
    return(bassline)
```
